```python
import pandas as pd
```

```python
col0=["user_id","movie_id","rating"]
data0=pd.read_csv("u.data",sep="\t",names=col0,usecols=range(3))
data0
```

```python
col1=["movie_id","title"]
data1=pd.read_csv("u.item", sep="|",names=col1, usecols=range(2),encoding="ISO-8859-1")
data1
```

```python
data=pd.merge(data0,data1)
data
```

```python
user_rating=data.pivot_table(index="user_id",columns="title",values="rating")
user_rating
```

```python
corr_matrix=user_rating.corr(method="pearson",min_periods=100)
corr_matrix
```

```python
test_user=user_rating.iloc[0].dropna()
test_user
```

```python
initial_recommendation=pd.Series(dtype="float64")

for i in range (len(test_user.index)):
    sim=corr_matrix[test_user.index[i]].dropna()
    score=sim*test_user[i]
    initial_recommendation=initial_recommendation.append(score)

initial_recommendation.sort_values(inplace=True,ascending=False)
initial_recommendation
```

```python
initial_recommendation.sort_values(inplace=True,ascending=False)
```

```python
recommendation=initial_recommendation.groupby(initial_recommendation.index).sum()
recommendation=recommendation.drop(test_user.index)
recommendation.sort_values(inplace=True,ascending=False)
recommendation
```

```python
final_recommendation=[]

for i in range (len(recommendation.index)):
    if recommendation[i]>=3:
        final_recommendation.append(recommendation.index[i])
    
final_recommendation
```
